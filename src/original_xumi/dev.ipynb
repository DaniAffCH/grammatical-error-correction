```python
df = pd.read_csv("../../../wikisent2.txt", delimiter='\t')
df_shuffled = df.sample(frac=1, random_state=42)

# Split the dataset into train and test sets
train_ratio = 0.8  # You can adjust the ratio based on your preference
train_size = int(len(df_shuffled) * train_ratio)

train_set = df_shuffled[:train_size]
test_set = df_shuffled[train_size:]

# Save the shuffled and split datasets to new CSV files
train_set.to_csv('train_set.csv', index=False)
test_set.to_csv('test_set.csv', index=False)
```
